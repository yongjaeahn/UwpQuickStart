	sympy/scipy 비선형 연립방정식

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')

f1 = sp.Eq(2*x - y,sp.exp(-x))
f2 = sp.Eq(-x + 2*y,sp.exp(-y))

solutions = sp.solve((f1,f2))
print('연립방정식의 해 :', solutions)
for key, value in solutions.items():
  print('연립방정식의 해 :', key, eval(str(value)))
```

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')

f1 = sp.Eq(2*x - y,sp.exp(-x))
f2 = sp.Eq(-x + 2*y,sp.exp(-y))

solution = sp.nsolve((f1,f2), (x,y), (-5,-5))
print("x의 해 :",solution[0],"y의 해 : ",solution[1])
solutionX, solutionY = sp.nsolve((f1,f2), (x,y), (5,5))
print("x의 해 :",solutionX,"y의 해 : ",solutionY)
```

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')

f1 = 2*x - y - sp.exp(-x)
f2 = -x + 2*y - sp.exp(-y)

sp.plotting.plot3d(f1,xlabel='x',ylabel='y')
sp.plotting.plot3d(f2,xlabel='x',ylabel='y')
sp.plotting.plot3d(f1,f2,xlabel='x',ylabel='y')
```

```python
import math
from scipy import optimize

def f(x):
  return (2*x[0] - x[1] - math.exp(-x[0]), -x[0] + 2*x[1] - math.exp(-x[1]))

solutionX, solutionY = optimize.fsolve(f,(5,5))
print("x의 해 :",solutionX,"y의 해 : ",solutionY)
```

```python
import math
from scipy import optimize

def f(symbols):
  x,y = symbols
  return (2*x - y - math.exp(-x), -x + 2*y - math.exp(-y))

solutionX, solutionY = optimize.fsolve(f,(-5,-5))
print("x의 해 :",solutionX,"y의 해 : ",solutionY)
```
